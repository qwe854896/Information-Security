# random walk algorithm

```python
n_stories = 1000
t_max = 200
```

```python
t = np.arange(t_max)
np.random.randint(0, 1+1, (n_stories, t_max))
steps = 2 * np.random.randint(0, 1+1, (n_stories, t_max)) - 1
np.unique(steps)
```

```python
positions = np.cumsum(steps, axis=1)
sq_distance = positions**2
```

```python
mean_sq_distance = np.mean(sq_distance, axis=0)
```

```python
plt.figure(figsize=(4, 3))
plt.plot(t ,np.sqrt(mean_sq_distance), 'g.', np.sqrt(t), 'y-')
plt.xlabel(r"$t$")
plt.ylabel(r"$\sqrt{\langle (\delta x)^2 \rangle}$")
plt.tight_layout()
```

## Regularization

### fitting a nth order polynomial

```python
rng = np.random.RandomState(0)
x = 2 * rng.rand(100) - 1

f = lambda t: 1.2 * t**2 + .1 * t**3 - .4 * t**5 - .5 * t**9
y = f(x) + .4 * rng.normal(size=100)

x_test = np.linspace(-1, 1, 100)

plt.figure(figsize=(6, 4))
plt.scatter(x, y, s=4)
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x, y, s=4)

X = np.array([x**i for i in range(5)]).T
X_test = np.array([x_test**i for i in range(5)]).T
regr = LinearRegression()
regr.fit(X, y)
plt.plot(x_test, regr.predict(X_test), label='4th order')

X = np.array([x**i for i in range(10)]).T
X_test = np.array([x_test**i for i in range(10)]).T
regr = LinearRegression()
regr.fit(X, y)
plt.plot(x_test, regr.predict(X_test), label='9th order')

#plt.plot(x_test, f(x_test), label='Ground truth')

plt.legend(loc='best')
plt.axis('tight')
plt.title('Fitting a 4th and a 9th order polynomial')
```

### Classification of Handwritten Digits

#### Data

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
fig = plt.figure(figsize=(6, 6))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    ax.text(0, 7, str(digits.target[i]))
```

#### Plot a Projection

```python
plt.figure()

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
proj = pca.fit_transform(digits.data)
plt.scatter(proj[:, 0], proj[:, 1], c=digits.target, cmap='Paired')
plt.colorbar()
```

#### Gaussian NB

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target)

cls = GaussianNB()
cls.fit(X_train, y_train)

predicted = cls.predict(X_test)
expected = y_test

fig = plt.figure(figsize=(6, 6))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(X_test.reshape(-1, 8, 8)[i], cmap=plt.cm.binary, interpolation='nearest')
    if predicted[i] == expected[i]:
        ax.text(0, 7, str(predicted[i]), color='green')
    else:
        ax.text(0, 7, str(predicted[i]), color='red')
```

#### Results

```python
matches = (predicted == expected)
print('Correction: ', matches.sum())
print('All: ', len(matches))
print('Correct rate: ', matches.sum() / float(len(matches)))
```

```python
from sklearn import metrics
print(metrics.classification_report(expected, predicted))
```

```python
print(metrics.confusion_matrix(expected, predicted))
```
