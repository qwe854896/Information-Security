```
%matplotlib inline
import matplotlib.pyplot as plt
```

```
x = np.arange(-1, 2, 0.1)
y = x ** 2 - 2 * x + 1
plt.plot(x, y)
```

```
x = np.arange(-6, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = .04 * x ** 2 + .08 * x - .06

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos')
plt.plot(x, y3, linestyle=':', label='test')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Some Function')
plt.legend()
plt.show()
```
